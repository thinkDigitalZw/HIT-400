```python
#train the model straight away
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

#initialize the Linear regression model and set to variable dataModel
dataModel = LinearRegression()

#fit or train the model
dataModel.fit(dataSet[['age', 'anaemia', 'creatinine_phosphokinase', 'diabetes', 'ejection_fraction', 'high_blood_pressure', 'serum_creatinine', 'serum_sodium', 'sex', 'smoking', 'time']], dataSet['DEATH_EVENT'] )
```

```python
# Check the model's score
dataModel.score(dataSet[['age', 'anaemia', 'creatinine_phosphokinase', 'diabetes', 'ejection_fraction', 'high_blood_pressure', 'serum_creatinine', 'serum_sodium', 'sex', 'smoking', 'time']], dataSet['DEATH_EVENT'])
```
